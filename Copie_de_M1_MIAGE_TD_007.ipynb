```
#Representez dans une carte du monde les lieux de tournages par film (la taille de la bulle sera determin√©e par la popularit√© du lieu Ref. Question prec√©dente)
from geopy.geocoders import Nominatim
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim

_007['Shooting_Loc'] = _007['Shooting_Loc'].str.split(',')
_007 = _007.explode('Shooting_Loc')
geolocator = Nominatim(user_agent="myGeocoder")
_007['location_coords'] = _007['Shooting_Loc'].apply(geolocator.geocode)
_007['latitude'] = _007['location_coords'].apply(lambda x: x.latitude)
_007['longitude'] = _007['location_coords'].apply(lambda x: x.longitude)
plt.figure(figsize=(20, 16))
m = Basemap(projection='mill',llcrnrlat=-90,urcrnrlat=90,\
            llcrnrlon=-180,urcrnrlon=180,resolution='c')
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='white',lake_color='aqua')
m.drawmapboundary(fill_color='aqua')

_007['films_count'] = _007.groupby(['Shooting_Loc'])['Shooting_Loc'].transform('count')

m.scatter(_007['longitude'], _007['latitude'], latlon=True,
          c=_007['films_count'], s=_007['films_count']*50, alpha=0.5)

plt.show()
```
