### Liste de dictionnaire

* √† partir d'un fichier  [csv](./src/to_do/td_04/netflix_titles.csv)

* en r√©sultat une liste de dicts qui doit contenir le titre de la s√©rie, la plateforme de streaming (l√©gal), le nombre d'√©pisodes de la s√©rie, l'ann√©e de sortie et une note /10.


````python
# Code here
FILE_PATH = "./tv_shows.csv"

def get_plateforme(liste_de_s√©rie) -> list:
    plateforme = []
    if liste_de_s√©rie[7] == 1:
        plateforme += "Netflix"
    if liste_de_s√©rie[8] == 1:
        plateforme += "Hulu"
    if liste_de_s√©rie[9] == 1:
        plateforme += "Prime vid√©o"
    if liste_de_s√©rie[10] == 1:
        plateforme += "Disney +"
    return plateforme

    

#TODO 1 : Ouvrir fichier csv
with open(FILE_PATH, 'r') as data_file:
    tv_shows = csv.reader(data_file)
    tv_shows_dict = {}
    idx = 0
    for tv_show in tv_shows:
        if idx == 0:
            idx += 1
            continue
        imdb_note = tv_show[5].split("/")[0]
        rotten_note = tv_show[6].split("/")[0]
        if not imdb_note == "" and not rotten_note == "":
            imdb_note = float(imdb_note)
            rotten_note = float(rotten_note)
            new_note = f"{(imdb_note + rotten_note)/2}/10"
        elif not imdb_note == "":
            new_note = f"{imdb_note}/10"
        elif not rotten_note == "":
            new_note = f"{rotten_note}/10"
        else:
            new_note == "Pas de notes"

        tv_shows_dict[idx] = {
            "Titre": tv_show[2],
            "Ann√©e de sortie": tv_show[3],
            "Plateforme": get_plateforme(tv_show),
            "Note": new_note,
        }

        idx+=1
# print(tv_shows_dict)
    

````

````python
# Constants
FILE_PATH = "./netflix_titles_propre.csv"


# Import data
def get_data_from_csv(file_path) -> list:
    with open(file_path) as data_file:
        movies_and_shows = data_file.readlines()
        movies_and_shows.pop(0)
        shows = []
        for movie_and_show in movies_and_shows:
            movie_and_show = movie_and_show.split("|")
            if movie_and_show[1] == "TV Show":
                shows.append(movie_and_show)
        shows_dict = {}
        for show in shows:
            shows_dict[show[2]] = {
                "date de sortie": show[7],
                "plateforme": "Netflix",
                "nombre de saisons": show[9],
                "note": f"{randint(1, 10)}/10",
            }
    return shows_dict


shows = get_data_from_csv(FILE_PATH)
print(shows)
````
