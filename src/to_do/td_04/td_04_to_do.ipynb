### Liste de dictionnaire

* Ã  partir d'un fichier  [csv](./td_04)

````python
# Code here
liste_series = []
dict_series = {}
with open("tv_shows.csv","r",encoding='utf8') as data:
    lines = data.readlines()
    for line in lines:
        element = line.split("|")
        liste_series.append(element)
    for i in range(1,len(liste_series)):
        plat = ""
        if liste_series[i][6] == "1":
            plat += "Netflix"
        if liste_series[i][7] == "1":
            plat += " Hulu"
        if liste_series[i][8] == "1":
            plat += " Prime Video"
        if liste_series[i][9] == "1":
            plat += " Dinsney +"
        dict_series[liste_series[i][0]]={"Titre" : liste_series[i][1], "Plateforme" : plat, "AnnÃ©e de sortie" : liste_series[i][2], "Note" : liste_series[i][4]}        #nombre d'Ã©pisodes pas dans le dictionnaire pcq info pas prÃ©sente dans le fichier csv
    print(dict_series)
````
