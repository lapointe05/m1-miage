```python
# Importer la librairie pandas
import pandas as pd
import numpy as np

# lecture du  dataset 
file_path = "./tv_shows.csv"
tv_data = pd.read_csv(file_path, delimiter='|')

#Afficher les 10 premieres lignes du dataset
print(tv_data.loc[:9,:])

#cr√©ation d'une plateforme "streaming_platform"
plateform = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']
#tv_data.columns
col = ['ID', 'Title', 'Year', 'Age', 'IMDb', 'Rotten Tomatoes','Type']
tv_data= pd.melt(tv_data, id_vars=col, value_vars=plateform, var_name='streaming_platform')

#Drop les colonnes 
#d√©ja drop avec la foncyion melt
#dropcolonnes = tv_data.drop(['Netflix','Hulu', 'Prime Video', 'Disney+'], axis=1)

#Le nombre de shows par plateforme

resultat = tv_data.groupby('streaming_platform')['value'].sum()
print(f"Le nombre de shows par {resultat}")

#divisons la note de la base 10 pour un d√©but
#print(tv_data[('IMDb')])
tv_data[['Note','Base']]=tv_data[('IMDb')].str.split('/',expand=True)

#la moyenne par plateforme de streaming
tv_data[('Note')] = tv_data[('Note')].astype(float)
imdb = tv_data.groupby('streaming_platform')['Note'].mean()
#print(f"La moyenenne par  {imdb}")

#les meilleurs shows 
tv_data.head()

```
